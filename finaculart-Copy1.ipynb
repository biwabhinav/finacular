```python
def insertVariblesIntoTable(date, price): #to add data to the table a function is created.
    try:
        connection = mysql.connector.connect(host='localhost',
                                             database='finacular',
                                             user='root',
                                             password='Sakshn2307')
        cursor = connection.cursor()
        mySql_insert_query = """INSERT INTO FINACULARDATA4 (date,  price) 
                                VALUES (%s, %s) """

        recordTuple = (date, price)
        cursor.execute(mySql_insert_query, recordTuple)
        connection.commit()
        print("Record inserted successfully into FINACULARDATA4 table")

    except mysql.connector.Error as error:
        print("Failed to insert into MySQL table {}".format(error))
    
    finally:
        if (connection.is_connected()):
            cursor.close()
            connection.close()
            print("MySQL connection is closed")

    
            
            
```
